## Task6

Write a pandas program to add, subtract, multiply and divide two pandas Series created from Python lists: [2, 4, 6, 8, 10] and [1, 3, 5, 7, 9]
Print the result of the operations.
Operations with pandas Series should be similar to operations with NumPy arrays.

```python
list1 = pd.Series([2, 4, 6, 8, 10])
list2 = pd.Series([1, 3, 5, 7, 9])
lists = list1 + list2
print("Add two Series:")
print(lists)
print("Subtract two Series:")
lists = list1 - list2
print(lists)
print("Multiply two Series:")
lists = list1 * list2
print(lists)
print("Divide Series1 by Series2:")
lists = list1 / list2
print(lists)
```

## Task8

```python
all_data = {ticker: web.get_data_yahoo(ticker)
           for ticker in ['AMZN','DELL']}
price = pd.DataFrame({ticker: data['Adj Close']
                     for ticker, data in all_data.items()})
price.head()
```

```python
price.mean()
```

```python
%matplotlib inline

mpl.rc('figure', figsize=(8, 7))

style.use('ggplot')

price.plot()
plt.legend()
```

```python
%matplotlib inline
plt.scatter(price.AMZN,price.DELL)
```

```python
pd.plotting.scatter_matrix(price, diagonal='kde', figsize=(10, 10));
```

```python
np.corrcoef(price.AMZN,price.DELL)
```

```python
print("Between Amazon and Dell stocks there is a correlation strong and positive of 0.79")
```
